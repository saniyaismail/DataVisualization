```python
import netCDF4
from netcdftime import num2date

# Open the NetCDF file
nc_file = netCDF4.Dataset(vs_nc_file, mode='r')

# Access the time variable
time_var = nc_file.variables['day']

# Convert time values to datetime objects
dates = num2date(time_var[:], time_var.units)

# Extract the day from the datetime objects
days = [date.day for date in dates]

print(days)
```
