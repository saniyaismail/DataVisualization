```python
# to check artist popularity, dropping songs repeated over different playlists
duplicate_records = df[df.duplicated(subset=['track_name', 'track_album_id'], keep=False)]
```

```python
duplicate_records
```

```python
duplicate_records[duplicate_records['track_album_id'] == '2oCs0DGTsRO98Gh5ZSl2Cx']
```

```python
df_no_duplicates = df.drop_duplicates(subset=['track_name', 'track_album_id'], keep=False)
```

```python
df_no_duplicates.shape
```

```python
df_no_duplicates.to_csv('spotify_songs_without_duplicates.csv')
```

```python
artist_popularity = df_no_duplicates.groupby('track_artist')['track_popularity'].mean().reset_index()
artist_popularity = artist_popularity.sort_values(by='track_popularity', ascending=False)
```

```python
artist_popularity.head(30)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Select top 10 artists for visualization (optional)
top_artists = artist_popularity.head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x='track_artist', y='track_popularity', data=top_artists, palette='viridis')
plt.xlabel('Artist')
plt.ylabel('Average Track Popularity')
plt.title('Top 10 Artists by Average Track Popularity')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
